# Predictive Maintenance RUL Demo

This notebook demonstrates how to work with the summary data for the LSTM-MHSA remaining useful life prediction project.

It covers:

- Loading benchmark metrics
- Comparing models using the combined metric
- Reviewing robustness under limited training trajectories
- Plotting a sample RUL prediction curve

The data files are summary/demo files based on values reported in the published paper.


```
import pandas as pd
import matplotlib.pyplot as plt

benchmark = pd.read_csv("../data/benchmark_metrics.csv")
robustness = pd.read_csv("../data/training_size_robustness.csv")
sample_rul = pd.read_csv("../data/sample_rul_predictions.csv")

benchmark
```

## Benchmark comparison

```
ax = benchmark.plot(
    x="model",
    y="combined_metric",
    kind="bar",
    figsize=(10, 5),
    rot=25,
    legend=False
)
ax.set_title("Benchmark comparison using combined metric")
ax.set_ylabel("Combined metric (lower is better)")
ax.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Robustness to limited training trajectories

```
for model in robustness["model"].unique():
    subset = robustness[robustness["model"] == model]
    plt.plot(subset["training_trajectories"], subset["combined_metric"], marker="o", label=model)

plt.title("Robustness to limited complete run-to-failure trajectories")
plt.xlabel("Number of complete training trajectories")
plt.ylabel("Combined metric (lower is better)")
plt.xticks([5, 10, 20])
plt.gca().invert_xaxis()
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

## Sample RUL prediction curve

```
plt.figure(figsize=(10, 5))
plt.plot(sample_rul["time_hours"], sample_rul["true_rul_hours"], label="True RUL", linewidth=2.5)
plt.plot(sample_rul["time_hours"], sample_rul["proposed_lstm_mhsa_rul_hours"], label="Proposed LSTM-MHSA", linewidth=2)
plt.plot(sample_rul["time_hours"], sample_rul["lstm_rul_hours"], label="LSTM", linestyle="--")
plt.plot(sample_rul["time_hours"], sample_rul["mhsa_rul_hours"], label="MHSA", linestyle=":")
plt.title("Sample RUL prediction curve")
plt.xlabel("Time (hours)")
plt.ylabel("Remaining useful life (hours)")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

## Key interpretation

The proposed LSTM-MHSA model provides strong RUL prediction performance and remains stable when fewer complete run-to-failure trajectories are available. This makes it relevant for high-reliability equipment, where failure histories are often scarce.
